## Метод k-ближайших соседей (KNN)
Объект присваивается тому классу, который является наиболее распространённым среди k соседей данного элемента, классы которых 
уже известны


```python
# Разделим сначала выборку на обучающую и тестовую
from sklearn.model_selection import train_test_split
# пусть обучающая выборка = 75%, тестовая - 25%
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 42)
```

```python
Y_train # просто для примера посмотрим на результат
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
# Непосредственно модель KNN
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 10) # количество соседей = 10, рандомно
knn.fit(X_train, Y_train)
```

```python
Y_train_predict = knn.predict(X_train) # предсказание на обучающей
Y_test_predict = knn.predict(X_test) # предсказание на тестовой выборке
err_test = 1 - knn.score(X_test, Y_test) # ошибка предсказания на тестовой
err_test
```

```python
err_train = 1 - knn.score(X_train, Y_train) # ошибка предсказания на обучающей
err_train
```

Видно, что разница в ошибке классификации в тестовой выборке, грубо говоря, на 4% выше, чем в обучающей. Я считаю, такую разницу можно назвать небольшой.

```python
# ПОДБОР ПАРАМЕТРОВ
from sklearn.model_selection import GridSearchCV
nnb = [1, 3, 5, 7, 9, 11, 15, 19, 25, 35, 42] # рандомно подобранные числа
knn = KNeighborsClassifier() # Классификатор, по умолчанию кол-во соседей = 5
grid = GridSearchCV(knn, param_grid = {'n_neighbors': nnb}, cv=10) # подбор соседей
```

```python
grid.fit(X_train, Y_train)
```

```python
best_cv_err = 1 - grid.best_score_ # считаем ошибку при наиболее подходящем кол-ве соседей
best_n_neighbors = grid.best_estimator_.n_neighbors # "вытаскиваем" это кол-во
print(best_cv_err, best_n_neighbors) # вывод ошибки и оптимальное кол-во соседей
```

```python
knn = KNeighborsClassifier(n_neighbors = best_n_neighbors).fit(X_train, Y_train)

err_train = 1 - knn.score(X_train, Y_train)
err_test = 1 - knn.score(X_test, Y_test)

print(err_train, err_test)
```

Удивительным образом ошибка в данной модели при k = 1 в обучающей выборке уменьшилась с 20% до 0% (по сравнению с k = 10), а в тестовой - с 24.2% до 21.3% 

```python
# проведем подбор оптимального кол-ва соседей еще и графическим способом
k = range(1, 55, 2)
list_err_train = []
list_err_test = []
for i in k:
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, Y_train)
    list_err_train.append(np.mean(Y_train != knn.predict(X_train)))
    list_err_test.append(np.mean(Y_test  != knn.predict(X_test)))
    print(i)
```

```python
plt.plot(k, list_err_test, label='test')
plt.plot(k, list_err_train, label='train')
plt.xlabel('k')
plt.ylabel('error_value')
plt.legend(loc='lower right')
plt.grid()
```

Очевидно из графика, что минимальная ошибка на тестовой и обучающей выборке достигается при количестве соседей  = 1.

```python
from sklearn.neural_network import MLPClassifier
# Обучение классификатора(данные ранее уже были разбиты на обучающ. и тест. выборки)
# Обучаем однослойную нейронную сеть (100 нейронов в скрытом слое). На выходе один нейрон, метод оптимизации - lbfgs
mlp_model = MLPClassifier(hidden_layer_sizes = (100,), solver = 'lbfgs',
                          activation = 'logistic', random_state = 42)
mlp_model.fit(X_train, Y_train)
```

```python
#  Подбор оптимального параметра регуляризации alpha
alpha_arr = np.logspace(-3, 2, 21)
test_err = []
train_err = []
for alpha in alpha_arr:
    mlp_model = MLPClassifier(alpha = alpha, hidden_layer_sizes = (100,),
                              solver = 'lbfgs', activation = 'logistic', random_state = 42)
    mlp_model.fit(X_train, Y_train)

    Y_train_pred = mlp_model.predict(X_train)
    Y_test_pred = mlp_model.predict(X_test)
    train_err.append(np.mean(Y_train != Y_train_pred))
    test_err.append(np.mean(Y_test != Y_test_pred))
```

### Random forests. 
Проведем классификацию по классам с помощью случайного леса. Основная идея заключается в использовании большого ансамбля решающих деревьев, каждое из которых само по себе даёт очень невысокое качество классификации, но за счёт их большого количества результат получается хорошим.

```python
from sklearn import ensemble
rf = ensemble.RandomForestClassifier(n_estimators = 1000, min_samples_split = 3) 
# первый арг. - кол-во деревьев,второй - кол-во разветвлений в дереве
rf.fit(X_train, Y_train)

err_train = 1 - rf.score(X_train, Y_train)
err_test = 1 - rf.score(X_test, Y_test)

print(err_train, err_test)
```

Ошибка в обучающей выборке снова равна нулю, а в тестовой на 35% меньше, чем в методе K-ближайших соседей, однако найденная точность совпадает с точностью использования нейронных сетей.

```python
#Значимость признаков (смотрится в модели Random Forest)
importances = rf.feature_importances_
indices = np.argsort(importances)[::-1]

print("Feature ranking:")

N, d = X_train.shape
feature_names = X.columns
for f in range(d):
    print("%2d. feature '%5s' (%f)" % (f + 1, feature_names[indices[f]], importances[indices[f]]))
```

Можно заметить, что самый большой вклад в модели сыграла высота, на втором месте - тип почвы, остальные - примерно в равных долях

```python
plt.title("Feature importances")
plt.bar(range(d), importances[indices[:d]], align = 'center', color = 'g')
plt.xticks(range(d), feature_names[indices[:d]], rotation = 90)
plt.xlim([-1, d])
```

Построим график зависимости ошибки на тестовой и обучающей выборках, а также oob-ошибки от количества деревьев. Тоесть проивезедм подбор оптимального количества деревьев

```python
trees_arr = np.arange(15, 150, 5)
test_err = []
train_err = []
oob_err = []
for trees in trees_arr:
    rf = ensemble.RandomForestClassifier(n_estimators = trees, oob_score=True)
    rf.fit(X_train, Y_train)
    train_err.append(np.mean(Y_train != rf.predict(X_train)))
    test_err.append(np.mean(Y_test  != rf.predict(X_test)))
    oob_err.append(1 - rf.oob_score_)
    
plt.plot(trees_arr, train_err, 'b-o', label = 'train')
plt.plot(trees_arr, test_err, 'r-o', label = 'test')
plt.plot(trees_arr, oob_err, 'g-o', label = 'oob')
plt.xlim([np.max(trees_arr), np.min(trees_arr)])
plt.title('Error vs. number of trees')
plt.xlabel('trees')
plt.ylabel('error')
plt.legend()
pass
```

На обучающей выборке очевидно некое постоянство, ошибка преимущественно = 0, на тестовой выборке наблюдаются незначительные колебания, график oob-ошибки возрастает при уменьшении количества деревьев с 70 до 0.

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)

print("Минимальная ошибка на обучающей выборке: ", min_train_err)
print("Минимальная ошибка на тестовой выборке: ", min_test_err)
```

```python
trees_opt = trees_arr[test_err == min_test_err]
print("Оптимальное количество деревьев: ", trees_opt)
```

```python
min_oob_err = np.min(oob_err)
trees_opt = trees_arr[oob_err == min_oob_err]
print(trees_opt) # при таком количестве деревьев достигается минимум  oob-ошибки
```

### Extremely Randomized Trees
Сверхслучайные деревья. Похожи на Случайные деревья, но разделения в узлах происходит шире. В экстремально случайных деревьях выбор признака, по которому поисходит ветвление выбирается абсолютно случайным образом.

```python
err_train = np.mean(Y_train != ert.predict(X_train))
err_test  = np.mean(Y_test  != ert.predict(X_test))

print('Ошибка на обучающей выборке = ',err_train)
print('Ошибка на тестовой выборке = ',err_test)
```

```python
#повторим
ert = ensemble.ExtraTreesClassifier(n_estimators = 100)
ert.fit(X_train, Y_train)
err_train = np.mean(Y_train != ert.predict(X_train))
err_test  = np.mean(Y_test  != ert.predict(X_test))

print('Ошибка на обучающей выборке = ',err_train)
print('Ошибка на тестовой выборке = ',err_test)
```

На данный момент этот алгоритм показывает неплохие результаты, относительно других. Посмотрим, можно ли улучшить показатели, подобрав оптимальное количество деревьев.

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)

print("Минимальная ошибка на обучающей выборке: ", min_train_err)
print("Минимальная ошибка на тестовой выборке: ", min_test_err)
```

```python
trees_opt = trees_arr[test_err == min_test_err]
print(trees_opt)
```

```python
min_oob_err = np.min(oob_err)
trees_opt = trees_arr[oob_err == min_oob_err]
print(trees_opt)
```

### AdaBoost
Каждый следующий классификатор строится по объектам, которые плохо классифицируются предыдущими классификаторами. AdaBoost вызывает слабый классификатор в цикле.

```python
ab = ensemble.AdaBoostClassifier(n_estimators = 100)
ab.fit(X_train, Y_train)
```

```python
err_train = np.mean(Y_train != ab.predict(X_train))
err_test  = np.mean(Y_test  != ab.predict(X_test))

print("Ошибка на обучающей выборке: ", err_train)
print("Ошибка на тестовой выборке: ", err_test)
```

Определенно, ошибка получилась очень большой, наблюдаем недообучение. Снова постараемся улучшить показатели.

```python
estim_arr = np.arange(10, 150, 5)
test_err = []
train_err = []
for estim in estim_arr:
    ab = ensemble.AdaBoostClassifier(n_estimators = estim)
    ab.fit(X_train, Y_train)
    train_err.append(np.mean(Y_train != ab.predict(X_train)))
    test_err.append(np.mean(Y_test  != ab.predict(X_test)))

plt.plot(estim_arr, train_err, 'b-o', label = 'train')
plt.plot(estim_arr, test_err, 'r-o', label = 'test')
plt.xlim([np.max(estim_arr), np.min(estim_arr)])
plt.title('Error vs. number of estimators')
plt.xlabel('estimators')
plt.ylabel('error')
plt.legend()
pass
```

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)

print("Минимальная ошибка на обучающей выборке: ", min_train_err)
print("Минимальная ошибка на тестовой выборке: ", min_test_err)
```

```python
trees_opt = estim_arr[test_err == min_test_err]
print(trees_opt)
```

### Gradient Boosted Trees
Градиентный бустинг является методом регрессии и классификации, который производит модель прогнозирования в виде ансамбля слабых моделей прогнозирования, как правило дерев решений . Он строит модель поэтапно, как и другие методы бустинга, обобщает их, позволяя оптимизировать произвольную дифференцируемую функцию потерь.

```python
gbt = ensemble.GradientBoostingClassifier(n_estimators = 100)
gbt.fit(X_train, Y_train)
```

```python
err_train = np.mean(Y_train != gbt.predict(X_train))
err_test  = np.mean(Y_test  != gbt.predict(X_test))

print('Ошибка на обучающей выборке:', err_train)
print('Ошибка на тестовой выборке:', err_test)
```

Видим, что результаты на тестовой выборке лучше, чем у метода KNN, но хуже, например, чем у нейронных сетей. Вновь попробуем улучшить показатели.

```python
trees_arr = np.arange(10, 150, 5)
test_err = []
train_err = []
for trees in trees_arr:
    gbt = ensemble.GradientBoostingClassifier(n_estimators = trees)
    gbt.fit(X_train, Y_train)
    train_err.append(np.mean(Y_train != gbt.predict(X_train)))
    test_err.append(np.mean(Y_test  != gbt.predict(X_test)))

plt.plot(trees_arr, train_err, 'b-o', label = 'train')
plt.plot(trees_arr, test_err, 'r-o', label = 'test')
plt.xlim([np.max(trees_arr), np.min(trees_arr)])
plt.title('Error vs. number of trees')
plt.xlabel('trees')
plt.ylabel('error')
plt.legend()
pass
```

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)

print("Минимальная ошибка на обучающей выборке: ", min_train_err)
print("Минимальная ошибка на тестовой выборке: ", min_test_err)
```

```python
trees_opt = trees_arr[test_err == min_test_err]
print(trees_opt)
```

В результате при подборе оптимального количеста деревьев удалось улучшить результат незначительно.

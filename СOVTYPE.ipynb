```python
data = data[0:6000] # сокращаем выборку
""" Нормализация количественных признаков ( 10 - 53 столбцы содержат бинарные признаки, их не нужно норм-ть). Нормализация 
необходима,чтобы масштабируемость данных не влияла на итоговый результат работы того или алгоритма. Например, если одна 
переменная измен-ся от 0 до 10000, а другая от 0 до 1."""
data_numerical = data.loc[:,0:9]
```

```python
data_numerical = (data_numerical - data_numerical.mean(axis = 0))/data_numerical.std(axis = 0) # Формула для кол.признаков:
# вычесть среднее и поделить на стандартное отклонение, чтобы среднее стало нулевым, а стандартное отклонение единицей
```

```python
data = data[0:6000] # на всей БД так и не запустился knn, поэтому сокращаем выборку

# Метод k-ближайших соседей (KNN)
"""Объект присваивается тому классу, который является наиболее распространённым среди k соседей данного элемента, классы которых 
уже известны."""
# Разделим сначала выборку на обучающую и тестовую
from sklearn.model_selection import train_test_split
# пусть обучающая выборка = 75%, тестовая - 25%
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 42) 
```

Видно, что разница в ошибке классификации в тестовой выборке, грубо говоря, на 3% выше, чем в обучающей. Я считаю, такую разницу можно назвать небольшой.

```python
# ПОДБОР ПАРАМЕТРОВ
from sklearn.model_selection import GridSearchCV
nnb = [1, 3, 5, 7, 9, 11, 15, 19, 25, 35, 42] # рандомно подобранные числа
knn = KNeighborsClassifier() # Классификатор, по умолчанию кол-во соседей = 5
grid = GridSearchCV(knn, param_grid = {'n_neighbors': nnb}, cv=10) # подбор соседей
```

```python
grid.fit(X_train, Y_train)
```

```python
best_cv_err = 1 - grid.best_score_ # считаем ошибку при наиболее подходящем кол-ве соседей
best_n_neighbors = grid.best_estimator_.n_neighbors # "вытаскиваем" это кол-во
print(best_cv_err, best_n_neighbors) # вывод ошибки и оптимальное кол-во соседей
```

```python
knn = KNeighborsClassifier(n_neighbors = best_n_neighbors).fit(X_train, Y_train)

err_train = 1 - knn.score(X_train, Y_train)
err_test = 1 - knn.score(X_test, Y_test)

print(err_train, err_test)
```

Удивительным образом ошибка в данной модели при k = 1 в обучающей выборке уменьшилась с 28,1% до 0% (по сравнению с k = 10), а в тестовой - с 31,3% до 19,3% (больше чем в полтора раза уменьшилась)

```python
# проведем подбор оптимального кол-ва соседей еще и графическим способом
k = range(1, 55, 2)
list_err_train = []
list_err_test = []
for i in k:
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, Y_train)
    list_err_train.append(np.mean(Y_train != knn.predict(X_train)))
    list_err_test.append(np.mean(Y_test  != knn.predict(X_test)))
    print(i)
```

```python
plt.plot(k, list_err_test, label='test')
plt.plot(k, list_err_train, label='train')
plt.xlabel('k')
plt.ylabel('error_value')
plt.legend(loc='lower right')
plt.grid()
```

Сложно интерпретировать данный график, поскольку речь идет не о бинарной классификации, но определенно видно, что при k=1 ошибка
классификации минимальна.

```python
#Random Forest - случайный лес. Основная идея заключается в использовании большого ансамбля решающих деревьев, каждое из которых
# само по себе даёт очень невысокое качество классификации, но за счёт их большого количества результат получается хорошим.
from sklearn import ensemble
rf = ensemble.RandomForestClassifier(n_estimators = 1000, min_samples_split = 3) # первый арг. - кол-во деревьев,второй - кол-во 
# разветвлений в дереве
rf.fit(X_train, Y_train)

err_train = 1 - rf.score(X_train, Y_train)
err_test = 1 - rf.score(X_test, Y_test)

print(err_train, err_test)
```

Ошибка в обучающей выборке снова равна нулю, а в тестовой на 1,4% меньше, чем в методе K-ближайших соседей

```python
#Значимость признаков (смотрится в модели Random Forest)
importances = rf.feature_importances_
indices = np.argsort(importances)[::-1]

print("Feature ranking:")

N, d = X_train.shape
feature_names = X.columns
for f in range(d):
    print("%2d. feature '%5s' (%f)" % (f + 1, feature_names[indices[f]], importances[indices[f]]))
```

Можно заметить, что самый большой вклад в модели сыграла высота, а вот 11 признаков внесли вклад приближенно равный нулю, в их числе, например, тип почвы.

```python
#Extremely Randomized Trees - Сверхслучайные деревья. Похожи на Случайные деревья, но разделения в узлах происходит шире
ert = ensemble.ExtraTreesClassifier(n_estimators = 100).fit(X_train, Y_train)

err_train = 1 - ert.score(X_train, Y_train)
err_test = 1 - ert.score(X_test, Y_test)

print(err_train, err_test)
```

В данном случае ошибка в обучающей выборке та же, а в тестовой нечто среднее между двумя предыдущими методами.

В итоге, из использованных трех методов лучший результат показал метод «случайный лес», а худший - KNN.
То, что ошибка в обучающей выборке равняется во всех трех методах нулю, я связываю с везением в выборке данных.
